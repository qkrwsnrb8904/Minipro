```python
import requests
from bs4 import BeautifulSoup
import csv

url = 'https://store.emart.com/branch/list.do?trcknCode=header_store'
headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36'}

response = requests.get(url, headers=headers)
soup = BeautifulSoup(response.content, 'html.parser')

stores = soup.find_all('div', class_='store-info')

store_data = []
for store in stores:
    name = store.find('h3', class_='store-name').text.strip()
    address = store.find('p', class_='store-address').text.strip()
    if '서울' in address:
        store_data.append({'name': name, 'address': address})

with open('emart_seoul_branches.csv', 'w', newline='', encoding='utf-8') as file:
    writer = csv.writer(file)
    writer.writerow(['Branch Name', 'Address'])
    for store in store_data:
        writer.writerow([store['name'], store['address']])

print("서울 이마트 매장 정보가 'emart_seoul_branches.csv' 파일에 저장되었습니다.")
```
